```python
import json
'''load data'''
```

```python
def load(filename):
    with open(filename,"r") as f:
        data=json.load(f)
        return data
```

```python
data=load("data.json")
```

```python
'''cleaning and structring'''
```

```python
cleaned=[]
uni=set()
```

```python
def clean(data):
    tn={"one":1,"two":2,"three":3,"four":4,"five":5,"true":1}
    for user in data:
        rrating=str(user["rating"]).strip().lower()
        if rrating in tn:
            rrating=tn[rrating]
            user["rating"]=rrating
        #nonevalues
        ager=user.get("age")
        if ager==None or ager=="":
            user["age"]="None"
   #dubli
        if user["name"] not in uni:
            uni.add(user["name"])
            cleaned.append(user)
    return cleaned
result=clean(data)
print(result)
            
```

```python
def insite(result):
    tr=0
    avg=0
    user=0
    for user in result:
        tr+=float(user["rating"])
        avg=tr/len(result)
    print(avg)
insite(result)
```

```python
def pr(result):
    pr=0
    pp=0
    for user in result:
        if (float(user["rating"]))<3:
            pr+=1
            pp=((pr/len(result))*100)
    print(pp)
pr(result)
```

```python
def rec(result):
    for user in result:
        if float(user["rating"])>3:
            print(user["name"],user["brand"])
        else:
            print(user["name"],"fellow")
rec(result)
```
