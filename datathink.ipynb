```python
import json
'''load data'''
```

```python
def load(filename):
    with open(filename,"r") as f:
        data=json.load(f)
        return data
```

```python
data=load("data.json")
```

```python
'''cleaning and structring'''
```

```python
cleaned=[]
uni=set()
```

```python
def clean(data):
    tn={"one":1,"two":2,"three":3,"four":4,"five":5,"true":1}
    for user in data:
        rrating=str(user["rating"]).strip().lower()
        if rrating in tn:
            rrating=tn[rrating]
            user["rating"]=rrating
        print(user["rating"])
        #nonevalues
        ager=user.get("age")
        if ager==None or ager=="":
            user["age"]="None"
        print(user["age"])
   #dubli
        if user["name"] not in uni:
            uni.add(user["name"])
            cleaned.append(user)
    return cleaned
result=clean(data)
print(result)
            
```
